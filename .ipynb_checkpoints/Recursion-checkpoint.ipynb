```python
# Recursion: A function calling itself.

# By Default the limit of the Recursion is 1000. But we can set a limit as much we need. It's great right.
```

```python
# Example for Recursion;

# Define a function.

import sys

sys.setrecursionlimit(100)
print(sys.getrecursionlimit())

i = 0

def Rec():
    
    global i
    i+=1
    
    print("Hello", i)
    
    Rec()
    
Rec()
```

```python
#Finding Factorial of the number using Recursion:

def fact(n):
    
    if n==0:
        
        return(1)
    
    return(n*fact(n-1))
    

n = int(input("Enter the value: "))
result = fact(n)
print(result)

```
