```python
#Anonymous functions (Lambda):

f = lambda a,b : a+b

print(f(2,3))
```

```python
#Filter:Filter will takes two arguments (Function and Iterable)

def is_even(n):
    
    return n%2==0

arr = [2,3,4,5,6,7,8,9]

evens = list(filter(is_even,arr))
print(evens)
    
    
```

```python
#Filter using Anonymous function Lambda:

is_even = lambda n : n%2==0

arr = [2,3,4,5,6,7,8,9]

evens = list(filter(is_even,arr))
print(evens)
```

```python
#Filter is somthing used filter the things based on the function. In above methods we first defined the function,
#and called it inside the filter. But in the following case we can use Anonymous function directly into the filter.

arr = [2,3,4,5,6,7,8,9]

#This time we seperate odd numbers. We need to deal it with the various cases right.

odds = list(filter(lambda n : n%2!=0, arr))
print(odds)

```

```python
#Map: Like filter, map also take two arguments (Function, Iterable), Map will change all the valuse as per our idea.

arr = [2,3,4,5,6,7]

doubles = list(map(lambda n : n*2, arr))
print(doubles)
```

```python
#Reduce: Which is used to reduce the value. Unlike map and filter reduce function belongs to a module functools.

from functools import reduce

arr = [2,3,4,5,6,7]

sum_of_arr = reduce(lambda a,b : a+b, arr)

print(sum_of_arr)
```
